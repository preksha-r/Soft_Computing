Demonstrate Crisp Set for Iris Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


iris = sns.load_dataset("iris")


iris['crisp_sepal_length'] = (iris['sepal_length'] >= 6.0).astype(int)
iris['crisp_sepal_width'] = (iris['sepal_width'] >= 3.0).astype(int)
iris['crisp_petal_length'] = (iris['petal_length'] >= 5.0).astype(int)
iris['crisp_petal_width'] = (iris['petal_width'] >= 1.5).astype(int)


print(iris.head())


print("\nCrisp Set Membership Counts:")
print("Sepal Length ≥ 6.0:\n", iris['crisp_sepal_length'].value_counts())
print("Sepal Width ≥ 3.0:\n", iris['crisp_sepal_width'].value_counts())
print("Petal Length ≥ 5.0:\n", iris['crisp_petal_length'].value_counts())
print("Petal Width ≥ 1.5:\n", iris['crisp_petal_width'].value_counts())


features = [
    ('sepal_length', 'crisp_sepal_length', 'Sepal Length ≥ 6.0'),
    ('sepal_width', 'crisp_sepal_width', 'Sepal Width ≥ 3.0'),
    ('petal_length', 'crisp_petal_length', 'Petal Length ≥ 5.0'),
    ('petal_width', 'crisp_petal_width', 'Petal Width ≥ 1.5'),
]

for feature, crisp_col, title in features:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=iris, x=feature, hue=crisp_col, palette={0: "gray", 1: "green"}, bins=20, multiple="stack")
    plt.title(f'Crisp Set Based on {title}')
    plt.xlabel(feature.replace("_", " ").title())
    plt.ylabel('Count')
    plt.grid(True)
    plt.legend(title='In Crisp Set')
    plt.show()
```
